```python
import langgraph
```

```python
from langgraph.graph import StateGraph
```

```python
graph=StateGraph(dict)
```

```python
def taskA(state):
  print('A started working')
  return(state)
def taskB(state):
  print('B started working')
  return(state)
def taskC(state):
  print('C started working')
  return(state)
```

```python
graph.add_node('A',taskA)
graph.add_node('B',taskB)
graph.add_node('C',taskC)
```

```python
graph.add_edge('A','B')
graph.add_edge('B','C')
```

```python
graph.set_entry_point('A')
```

```python
app=graph.compile()
app
```
